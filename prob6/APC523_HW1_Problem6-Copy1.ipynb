```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
arr = np.linspace(1.0, 10.0, 1001)
```

```python
sqarr = arr
NLOOPS = 52
for i in range(NLOOPS):
    sqarr = np.sqrt(sqarr)
for i in range(NLOOPS):
    sqarr = sqarr**2

plt.figure(figsize=(7,6))
plt.plot(arr, sqarr, label='{:d} successive sqrts and sqs'.format(NLOOPS))
plt.plot(arr, arr, linestyle='dashed', color='gray')
plt.legend(fontsize=12)
# plt.savefig('sqrt_52.png')
```

```python
uniq_sqarr = np.unique(sqarr)
uniq_sqarr
```

```python
res = uniq_sqarr
for i in range(NLOOPS):
    res = np.sqrt(res)
print([float.hex(r) for r in res])
```

```python
float.hex(res)
```

```python
res = uniq_sqarr[-3]
for i in range(NLOOPS):
    res = np.sqrt(res)
print(float.hex(res))
```

```python
for sq_res in uniq_sqarr:
    for i in range(NLOOPS):
        sq_res = np.sqrt(sq_res)
    print(float.hex(sq_res))
```
